```python
import pandas as pd
import numpy as np
```

```python
file_a = "HST_mbh_Current2024.csv"
file_b = "HST_CIV_Sulentic2007_HSLA2018_finalprops.csv"

# Read the spreadsheets into dataframes
df_a = pd.read_csv(file_a)
df_b = pd.read_csv(file_b)

```

```python
# Round RA and DEC in both dataframes
df_a = df_a.replace(pd.NA, np.nan)
df_b = df_b.replace(pd.NA, np.nan)
df_a['RA_rounded'] = df_a['RA_deg'].round(1)#.astype(int)
df_a['DEC_rounded'] = df_a['Dec_deg'].round(1)#.astype(int)
df_b['RA_rounded'] = df_b['_RA'].round(1)#.astype(int)
df_b['DEC_rounded'] = df_b['_DE'].round(1)#.astype(int)

# print(df_a['RA_rounded'])
# print(df_b['RA_rounded'])
df_a_ra = df_a['RA_rounded'].to_numpy()
df_a_ra = df_a_ra[~np.isnan(df_a_ra)]
df_a_dec = df_a['DEC_rounded'].to_numpy()
df_a_dec = df_a_dec[~np.isnan(df_a_dec)]

df_b_ra = df_b['RA_rounded'].to_numpy()
df_b_dec = df_b['DEC_rounded'].to_numpy()

print(df_a_ra)
print(df_b_ra)
```

```python
matches = []
for index, val in enumerate(df_b_ra):
    ra_matches = np.where(df_a_ra == val)
    dec_matches = np.where(df_a_dec == df_b_dec[index])
    common_matches = np.intersect1d(ra_matches, dec_matches)
    matches.append([index, common_matches.tolist()])
```

```python
import glob
from astropy.io import fits

#dat = pd.read_csv("/Users/trevormccaffrey/Dropbox/HST/final_tables/HST_CIV_Sulentic2007_HSLA2018_finalprops.csv")
dat = pd.read_csv("HST_CIV_Sulentic2007_HSLA2018_finalprops.csv")
#print(dat.head())

names = dat["Final_Name"].values
names_spec = dat["Spec_Name"].values
inst_final = dat["Inst_final"].values

for i in range(len(names)):
    
    names[i] = "".join(names[i].split(" "))

path = "/RebinnedSpec_2022Aug11/"

wave_start = []
wave_end = []

for name_final, name,inst,i in zip(names, names_spec,inst_final,range(len(dat))):
    
    fn_list = glob.glob("RebinnedSpec_2022Aug11/%s*%s.fits"%(name,inst))
    #print(fn_list) 
    if len(fn_list)==0:
        continue
    else:
        fn = fn_list[0]
    spec = fits.open(fn)
    wave_orig = spec[1].data["Rest-Frame Wavelength"] #* ((1+.6953)/(1+.6992))
    flux_orig = spec[1].data["Coadded Flux (Arbitrary Units)"]
    z         = spec[1].data["Redshift"][0]
    flux_orig /= np.nanmedian(flux_orig)
    errs_orig = spec[1].data["Coadded Flux Errors"] / np.nanmedian(flux_orig)

    wave_start.append(wave_orig[0])
    wave_end.append(wave_orig[-1])
    print([wave_orig[0], wave_orig[-1]])
```

```python
wave_start = np.array(wave_start)
wave_end = np.array(wave_end)

wavelength_coverage = wave_end - wave_start

low = np.percentile(wavelength_coverage, 100 / 3)
med = np.percentile(wavelength_coverage, 200 / 3)
```

```python
wavelength_coverage.min(), low, med, wavelength_coverage.max()
```

```python
df = pd.read_csv(file_b)

wave_start_final = np.zeros(464)
wave_end_final = np.zeros(464)
wave_coverage_final = np.zeros(464)
wave_coverage_level = np.zeros(464)

names_final = df["Final_Name"].to_numpy()

for index, name in enumerate(names_final):

    valid_object = name in matched_objects_final_names
    if valid_object:
        index_final = np.where(matched_objects_final_names == name)
        wave_start_final[index_final] = wave_start[index]
        wave_end_final[index_final] = wave_end[index]
        wave_coverage_final[index_final] = wavelength_coverage[index]

        if wavelength_coverage[index] <= low:
            wave_coverage_level[index_final] = 1
        elif wavelength_coverage[index] <= med:
            wave_coverage_level[index_final] = 2
        else:
            wave_coverage_level[index_final] = 3
    
```

```python
new_df = pd.DataFrame(zip(wave_start_final, wave_end_final, wave_coverage_final, wave_coverage_level), columns=['Wavelength Start','Wavelength End', 'Wavelength Coverage', 'Wavelength Coverage Level'])
old_df = pd.read_csv('HST_mbh_Current2024New.csv')


new_df_final = pd.concat([old_df, new_df], axis=1)
new_df_final.to_csv('HST_mbh_Current2024New2.csv')
new_df_final.head(20)
```
